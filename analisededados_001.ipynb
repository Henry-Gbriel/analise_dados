Analise de dados


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales_data.csv")
print(df.head(3)) #Só as três primeiras linhas do dataframe
```

```python
print(df.head())
```

```python
media_preco = df["Preço_Unitário"].mean() #media

print(f"Essa é a média de valores {media_preco :.2f}")

mediana_preco = df["Preço_Unitário"].median() #mediana

print(f"Essa é a mediana de valores vendidos {mediana_preco :.2f}")
```

```python
#setor que mais vendeu
setor_mais= df.groupby("Categoria_Produto")["Quantidade_Vendida"].sum()

setor_t =  setor_mais.idxmax()
quantidade_t = setor_mais.max()

setor_m = setor_mais.idxmin()
quantidade_m = setor_mais.min()

print(setor_mais)

print(f"O setor que mais vendeu foi: {setor_t} com {quantidade_t} produtos vendidos.")

print(f"O setor que veio ter menor venda foi: {setor_m} com {quantidade_m} produtos vendidos. ")
```

```python
# Criar o gráfico de vendas por categoria
plt.figure(figsize=(8, 5))
df.groupby("Categoria_Produto")["Quantidade_Vendida"].sum().plot(kind="bar", color="royalblue")
plt.ylabel("Quantidade Vendida")
plt.title("Vendas por Categoria de Produto")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Criar o gráfico de vendas por cidade
plt.figure(figsize=(8, 5))
df.groupby("Cidade")["Quantidade_Vendida"].sum().plot(kind="bar", color="darkorange")
plt.ylabel("Quantidade Vendida")
plt.title("Vendas por Cidade")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
